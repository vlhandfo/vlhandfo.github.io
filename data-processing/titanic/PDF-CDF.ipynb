# Representing Probabilities Aboard the Titanic

[Demo Resource](https://demonstrations.wolfram.com/ConnectingTheCDFAndThePDF/)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/titanic.csv")

age_list = df["Age"].dropna()
mean_age = age_list.mean()

d = age_list.describe()

metrics_df = pd.DataFrame(d).T
metrics_df
```

```python
sns.displot(age_list, bins=16)
plt.title("Distribution of Age of Passengers Aboard the Titanic")
plt.axvline(x=mean_age, color="black", linestyle="dashed", label="Mean Age")
plt.legend()
plt.show()
```

```python
sns.displot(age_list, bins=16)
plt.title("Distribution of Age of Passengers Aboard the Titanic")
plt.axvline(x=age_list.median(), color="red", label="Median")
plt.axvline(x=d["75%"], color="red", linestyle="dashed", label="Q1")
plt.axvline(x=d["25%"], color="red", linestyle="dashed", label="Q3")
plt.legend()
plt.show()
```

## Probability Density Function (PDF)

```python
sns.displot(age_list, kind="kde")
plt.text(4, 0.035, "Probability Density Function (PDF)")
plt.axvline(x=mean_age, color="black", linestyle="dashed", label="Mean Age")
plt.axvline(x=16, color="red", label="Age 16")
plt.title("Distribution of Age of Passengers Aboard the Titanic")
plt.legend()
plt.show()
```

## Cumulative Distribution Function

```python
sns.displot(age_list, kind="ecdf")
plt.text(4.5, 1.075, "Empirical Culumative Distribution Function (PDF)")
plt.title("Distribution of Age of Passengers Aboard the Titanic")
plt.axvline(x=mean_age, color="black", linestyle="dashed", label="Mean Age")
plt.axvline(x=16, color="red", label="Age 16")
plt.legend()
plt.show()
```
